```python
import pandas as pd

data = pd.read_csv('politifact.csv', index_col=0)
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

data["ruling"] = label_encoder.fit_transform(data["ruling"])
data["name"] = label_encoder.fit_transform(data["name"])
data["quote"] = label_encoder.fit_transform(data["quote"])
data["description"] = label_encoder.fit_transform(data["description"])
data["month"] = label_encoder.fit_transform(data["month"])

data.head()
```

```python
from sklearn.model_selection import train_test_split

x = data.drop("ruling", axis=1)
y = data["ruling"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42,shuffle=True)
```

```python
import tensorflow as tf

model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(x_train.shape[1],)), 
    tf.keras.layers.Dense(32, activation='relu'),                                   
    tf.keras.layers.Dense(1)                                                       
])

model.compile(optimizer='adam', loss='mse', metrics=['mae'])

history = model.fit(x_train, y_train, epochs=100, validation_data=(x_test, y_test))

test_loss, test_mae = model.evaluate(x_test, y_test)
print(f"Test Loss: {test_loss}, Test MAE: {test_mae}")

predictions = model.predict(x_test)
print(predictions[:5]) 
y_test.head()
```
